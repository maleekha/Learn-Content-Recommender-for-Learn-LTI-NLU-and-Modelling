```python
stop_words = set(stopwords.words('english')) 
def remove_stop_words(example_sent):
    word_tokens = example_sent.lower().split(' ')   
    filtered_sentence = [w for w in word_tokens if not w in stop_words]   
    filtered_sentence = []   
    for w in word_tokens: 
        if w not in stop_words: 
            filtered_sentence.append(w)
    return ' '.join(filtered_sentence)

all_desc = data['title_and_summary'].values
all_pp = [remove_stop_words(s) for s in all_desc]
```

```python
all_pp
```

```python
replace_invalid = {
    '-':' ',
    r'\u202': '',
    '(':'',
    ')':'',
    '/':' ',
    "'s":'',
    ',':'',
    '. ':' ',
    '`':'',
    "\'":'',
    ':':'',
    ';':'',
    '?':'',
    '--':''
}
```

```python
def replace_end_fullstop(sent):
    if(sent[-1]=='.'):
        return sent[:-1]
    return sent

pre_processed = []
for sent in all_pp:
    for k in list(replace_invalid.keys()):
        sent = sent.replace(k, replace_invalid[k])
    pre_processed.append(replace_end_fullstop(sent))
    print(replace_end_fullstop(sent))
```

```python
pre_processed
```

```python
data['preprocessed'] = pre_processed
```

```python
data.to_csv('pre_processed_data.csv', index=False, header=True)
```

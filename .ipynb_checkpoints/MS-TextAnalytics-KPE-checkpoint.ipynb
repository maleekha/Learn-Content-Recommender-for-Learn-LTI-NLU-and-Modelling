```python
key = ""
endpoint = ""
```

```python
from azure.ai.textanalytics import TextAnalyticsClient
from azure.core.credentials import AzureKeyCredential

def authenticate_client():
    ta_credential = AzureKeyCredential(key)
    text_analytics_client = TextAnalyticsClient(
            endpoint=endpoint, 
            credential=ta_credential)
    return text_analytics_client

client = authenticate_client()
```

```python
import pandas as pd
import os

data = pd.read_csv("data.csv").title_and_summary.values
```

```python
key_phrases_data = []
```

```python
def key_phrase_extraction_example(client):

    for i in range(data.shape[0]):
        try:
            documents = [data[i]] # ["My cat might need to see a veterinarian."]

            response = client.extract_key_phrases(documents = documents)[0]

            if not response.is_error:
                print("\tKey Phrases:", response.key_phrases)
                key_phrases_data.append(response.key_phrases)
            else:
                print(response.id, response.error)

        except Exception as err:
            print("Encountered exception. {}".format(err))
        
key_phrase_extraction_example(client)
```

```python
df = pd.read_csv("data.csv")
df['key_phrases'] = key_phrases_data
df.to_csv("data_kpe.csv", header=True, index=False)
```
